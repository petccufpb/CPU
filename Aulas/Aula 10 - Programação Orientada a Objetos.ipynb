# **Programação Orientada a Objetos**

A Programação Orientada a Objetos é uma forma de programação baseada na aproximação entre o mundo real e o mundo virtual.  É uma maneira de programar que fornece um meio de estruturar programas a partir de **objetos**, os quais irão possuir características e comportamentos.

Essa nova forma de programar se baseia em 4 conceitos principais: **classes**, **objetos**, **atributos** e **métodos**.

## **O que são classes e métodos?**

Imagine que você comprou um carro recentemente e decide modelar esse carro usando programação orientada a objetos. O seu carro tem as **características** que você estava procurando: um motor 2.0 híbrido, azul escuro, quatro portas, câmbio automático etc. Ele também possui **comportamentos** que, provavelmente, foram o motivo de sua compra, como acelerar, desacelerar, acender os faróis, buzinar e tocar música. Podemos dizer que o carro novo é um **objeto**, onde suas características são seus **atributos** (dados atrelados ao objeto) e seus comportamentos são ações ou **métodos**.

Na loja onde você comprou o seu carro existiam vários outros muito similares, com quatro rodas, volante, câmbio, retrovisores, faróis, dentre outras partes. Observe que, apesar do seu carro ser único (por exemplo, possui um registro único no Departamento de Trânsito), podem existir outros carros com exatamente os mesmos atributos, outros apenas com alguns atributos similares e outros totalmente diferentes, mas que ainda assim são considerados carros. Podemos dizer então que seu objeto (carro novo) pode ser classificado (isto é, seu objeto pertence à uma classe) como um carro, e que seu carro nada mais é que uma instância dessa classe chamada "carro".

class-analogy (1).png

Assim, abstraindo um pouco a analogia, uma classe é um conjunto de características e comportamentos que definem o conjunto de objetos pertencentes à essa classe. Repare que a classe em si é um conceito abstrato, como um molde, que se torna concreto e palpável através da criação de um objeto. Chamamos essa criação de instanciação da classe, como se estivéssemos usando esse molde (classe) para criar um objeto.

## **Como criar uma classe?**

image.png

Todas as definições de classe começam com a palavra-chave `class`, que é seguida pelo nome da classe e dois pontos. Qualquer código que esteja recuado abaixo da definição da classe é considerado parte do corpo da classe. 

```python
class Cachorro:
  pass
```

O corpo da classe Cachorro consiste em uma única instrução: a palavra-chave `pass`. O `pass` costuma ser usado como um marcador de posição que indica para onde o código irá eventualmente. Ele permite que você execute este código sem que o Python gere um erro. 

**Atenção:** Os nomes das classes em Python são escritos na notação CapitalizedWords por convenção. Por exemplo, o nome de uma classe para uma raça específica de cachorro, como Jack Russell Terrier, seria escrita como JackRussellTerrier. 

## **Como criar objetos a partir de uma classe?**

A criação de um novo objeto a partir de uma classe é chamada de instanciar um objeto. Você pode instanciar um novo objeto Cachorro digitando o nome da classe, seguido de abrir e fechar parênteses: 

```python
Cachorro()
```

Agora você tem um novo objeto Cachorro em 0x7fb476cc4510. Esta sequência engraçada de letras e números é um endereço de memória que indica onde o objeto Cachorro está armazenado na memória do seu computador

A nova instância Cachorro está localizada em um endereço de memória diferente. Isso ocorre porque é uma instância inteiramente nova:

```python
a = Cachorro()
print(a)
print(type(a))
```

```python
b = Cachorro()
print(b)
print(type(b))
```

```python
a == b
```

Nós códigos acima, você cria dois novos objetos Cachorro e os atribui às variáveis a e b. Quando você compara a e b usando o operador ==, o resultado é False. Embora a e b sejam instâncias da classe Cachorro, eles representam dois objetos distintos na memória. 

## **Como criar atributos para as instâncias da classe?**

**Atributos das instâncias**

A classe Cachorro até agora não possui nenhum método ou atributo definido. Existem várias propriedades que podemos escolher representar, incluindo nome, idade, cor do pêlo e raça.

As propriedades que todos os objetos Cachorro devem ter são definidas em um método chamado **.__ init __ ()**. Cada vez que um novo objeto Cachorro é criado, .__ init __ () define o estado inicial do objeto atribuindo os valores das propriedades do objeto. Ou seja, .__ init __ () inicializa cada nova instância da classe. 

```python
class Cachorro:
  def __init__(self):
    self.nome = "Marley"
    self.idade = 2
    self.cor_pelo = "Dourado"
    self.raca = "Labrador"
```

No corpo do **.__ init __ ()**, existem duas declarações usando a variável `self`:

`self.nome = "Marley"`: cria um atributo chamado nome e atribui a ele o valor "Marley".

`self.idade = 2`: cria um atributo chamado idade e atribui a ele o valor 53.

**O que é *self*?**

Quando uma nova instância de classe é criada, a instância é automaticamente passada para o parâmetro self em **.__ init __ ()** para que novos atributos possam ser definidos no objeto. 


**Criando um objeto:**

Instanciando um objeto da classe Cachorro

```python
bulldog = Cachorro()
```

Acessando o atributo **nome**:

```python
bulldog.nome
```

Acessando um atributo que não existe:

```python
bulldog.temperamento
```

Alterando o nome de um atributo

```python
bulldog.nome = "Sirius"
print(bulldog.nome)
```

## **Passando parâmetros para o método __init__**

Além da variável **self** também podemos passar outros parâmetros para o **método __init__**:

```python
class Cachorro:
  def __init__(self, nome, idade, cor_pelo, raca):
    self.nome = nome 
    self.idade = idade
    self.cor_pelo = cor_pelo
    self.raca = raca
```

**Criando um objeto:**

Instanciando objetos e passando como parâmetro os valores que serão preenchinos nos atributos:

```python
python = Cachorro("Mike", 3, "Preto", "pug")
```

Acessando os valores dos atributos:

```python
python.nome
```

```python
python.raca
```

Instanciando outro objeto Cachorro:

```python
java = Cachorro("Java", 3, "Dourado", "Labrador")
```

```python
java.nome
```

```python
java.raca
```

## **Como criar atributos da classe?**

Atributos de classe são atributos que têm o mesmo valor para todas as instâncias da classe. Você pode definir um atributo da classe atribuindo um valor a um nome de variável fora do **.__ init __ ().**

Por exemplo, a seguinte classe Cachorro tem um atributo de classe chamado especie com o valor "Canis familiaris": 

```python
class Cachorro:
  especie = "Canis familiaris"

  def __init__(self, nome, idade, cor_pelo, raca):
    self.nome = nome 
    self.idade = idade
    self.cor_pelo = cor_pelo
    self.raca = raca
```

Os **atributos de classe** são definidos diretamente abaixo da primeira linha do nome da classe e são recuados por quatro espaços (identação). Eles devem sempre receber um valor inicial. Quando uma instância da classe é criada, os atributos da classe são criados automaticamente e atribuídos a seus valores iniciais.

**Dica:** Use **atributos de classe** para definir propriedades que devem ter o mesmo valor para cada instância de classe. Use a**tributos de instância** (que são definidos dentro do **.__ init __ ()** para propriedades que variam de uma instância para outra. 

Instanciando um objeto Cachorro:

```python
mila = Cachorro("Mila", 2, "Branco", "Indefinida")
```

Acessando o atributo da instância **"cor_pelo"**:

```python
mila.cor_pelo
```

Acessando o atributo da classe **"especies"**:

```python
mila.especie
```

## **Como criar métodos para classe?**

```python
class Violao:

  tampo = 'laminado'

  def __init__(self, marca, cordas, numCordas, cor):
    self.marca = marca
    self.cordas = cordas
    self.numCordas = numCordas
    self.cor = cor
    self.afinado = False
  
  def afinar(self):
    self.afinado = True

  def escala(self):
    if self.afinado:
      print("DÓ RÉ MI FÁ SOL LÁ SI")
    else:
      print("No peito dos Desafinados também bate um coração")
      
  def acorde(self, chord):
    print(f"{chord.upper()}")
	 
	 
	 
presente = Violao("Rozinni", "nylon", 7, "natural") 
presente.acorde('c')
presente.escala()
presente.tampo
presente.afinar()
```
