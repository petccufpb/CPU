Durante o curso, ja utilizamos funções sem saber o que seria uma função. Podemos definir uma função como um conjunto de comandos definidos em um módulo que executa uma atividade específica.

**Função não é método!**

Temos que tomar cuidado pra não confundir uma função com um método. O método determina o comportamento de um objeto (conceito que vocês vão aprender mais pra frente). Os métodos são chamados depois de um ponto "." exemplo: frase.split(). Já as funções são um conjunto de comandos que executam uma atividade específica em um módulo.

As funções possuem "parâmetros" que são o conteúdo que fica entre os parênteses quando chamamos uma função. Não existe um número limite pra quantidade de parâmetros que podemos passar em uma função.

## Funções com retorno

Podemos definir uma função que possua um retorno. Nesse caso precisamos apenas colocar o nome "return" e logo após o que desejamos retornar para o local no código onde a função foi chamada

**Exemplo:**

```python
def sub(a,b):
  return a-b
```

```python
resultado = sub(3,4)
print(resultado)
```

```python
print(f"{sub(3,4)}")
```

Nem toda função precisa ter um retorno. Algumas funções existem apenas para modificar variáveis sem retornar nenhum tipo de dado.

**Par ou impar**

Podemos definir uma função que receba um valor e retorne pra o usuário se o valor é par ou impar.

```python
def par_ou_impar(a):
  if a%2 == 0:
    return "par"
  else:
    return "ímpar"
```

```python
par_ou_impar(4)
```

```python
par_ou_impar(5)
```

Imagina ter que fazer essa sequência de comandos toda vez que precisamos saber se um número é par ou ímpar em um código, **seria muito custoso**. As funções existem justamente pra otimizar seu código e facilitar a vida do programador.

**Tirando a média entre 2 notas**

Podemos criar uma função que calcule a média aritmética entre duas notas

```python
def media(a, b):
  return (a+b)/2
```

```python
media(4.5,4.5)
```

Podemos chamar uma função dentro de outra função!

Por exemplo, se quisermos criar uma função que diga qual a maior palavra entre duas

```python
def maior(palavra1, palavra2):
  if len(palavra1) > len(palavra2):
    return palavra1
  else:
    return palavra2
```

```python
string1 = "Cigarro"
string2 = "Carro"
maior(string1, string2)
```

## Variaveis locais e globais


Quando usamos funções, começamos a trabalhar com variáveis internas ou locais e
com variáveis externas ou globais. A diferença entre elas é a visibilidade.

As variaveis criadas dentro de uma função são vistas apenas quando estamos na função, quando saímos do escopo da função essas variáveis deixam de existir no código. Já as variáveis globais são vistas a todo momento em qualquer parte do código.

```python
nome = "Augusto"
```

```python
def chamar():
  print(f"Venha comer {nome}!")
```

```python
chamar()
```

Nesse caso, a variável "nome" é global, pois mesmo que não tenhamos declado ela dentro da função conseguimos enxerga-la. Se a variavel for criada dentro da função, ela passa a ser uma variavel local e só consegue ser enxergada dentro da própria função, deixando de existir quando a função acaba.

**Exemplo:**

```python
def correr():
  primeiro_nome = "Ricardo"
  print(f"Venha correr {primeiro_nome}!")
```

```python
correr()
```

```python
print(f"Testando se {primeiro_nome} existe")
```

No caso de uma variável local ser criada com o mesmo nome de uma variável global, a variável que vai ser enxergada sempre é a local.

```python
carro = "uno amarelo"
```

```python
def veiculo():
  carro = "gol branco"
  print(f"Meu carro é um {carro}")
```

```python
veiculo()
```

```python
print(f"Meu carro é um {carro}")
```

## Parâmetros opcionais


De vez em quando, queremos produzir funções com certos valores padrões. Para isso, podemos utilizar parâmetros opcionais ao definir a função.

```python
##parâmetros opcionais
#potencia

#10² = 1*10*10
#10³ = 1*10*10*10

def potencia(numero, expoente=2):
  total = 1
  for i in range(expoente):
    total = total * numero
  print(f"{numero} elevado a {expoente} = {total}")

potencia(10, expoente=3)
```

```python
##parâmetros opcionais
##cabeçalho

def cabecalho(mensagem="CURSO DE PROGRAMAÇÃO PARA UNIVERSITÁRIOS"):
  print('='*len(mensagem))
  print(mensagem)
  print('='*len(mensagem))

cabecalho()
cabecalho("CPU")
```

##Desempacotamento


Em python, conseguimos receber *n* parâmetros usando o conceito de desempacotamento. Com um asterisco antes do nome do parâmetro, o Python entende que a função vai receber *n* valores, no caso do nosso exemplo. Ele vai salvar tudo em uma tupla. 

```python
##desempacotamento
##soma

def soma(*num):
  sum = 0
  for numero in num:
    sum += numero
  return sum

a = 15
print(a)
print(soma(1,3,5))
```

## Recursão

Recursão em programação é quando uma função chama ela mesma. Podemos comparar com o conceito de metalinguagem do português, quando a linguagem fala sobre a linguagem. Outra forma de pensar seria de relacionar com a seguinte imagem: 



mona.jpg

```python
##Função recursiva
#fatorial

#3! = 3*2*1 = 6
#4! = 4*3*2*1 = 4 * 3!  
#n!=n*(n-1)! 
#0! = 1

def fatorial(numero):
  print(f"Calculando o fatorial de {numero}")
  if numero == 1 or numero == 0:
    print(f"  Fatorial de {numero} = 1")
    return 1
  else:
    fat = numero*fatorial(numero-1)
    print(f"  Fatorial de {numero} = {fat}")
    return fat

print(fatorial(4))
```

Essa é uma forma de fazer uma função que calcula o fatorial recursivamente. Perceba que para encontrar o valor de `fatorial(4)` é necessário calcular `fatorial(3)` e para calcular o mesmo é necessário saber quanto é `fatorial(2)`.
